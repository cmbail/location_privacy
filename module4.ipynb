```python
class StringWrapper:
    def __init__(self, string):
        self.input = string
    def reversed(self):
        return(self.input[::-1])

wrapper = StringWrapper('Python')
print(wrapper.reversed())
```

```python
# Exercise 1: The "Digital Pet" Workshop
class Pet:
    def __init__(self, name):
        self.name = name
        self.hunger = 50
    def eat(self, food_amount):
        self.hunger -= food_amount
        if self.hunger < 0:
            self.hunger = 0
    def status(self):
        return f'My name is {self.name} and my hunger level is {self.hunger}.'

cat = Pet('Ruby')
print(cat)
print(cat.name)
print(cat.status())
cat.eat(20)
print(cat.status())
cat.eat(20)
print(cat.status())
cat.eat(20)
print(cat.status())
cat.eat(20)
```

```python
# Exercise 2: The "Hero & Wizard" Inheritance
class Hero:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return f"Hello, my name is {self.name}."

class Wizard(Hero):
    def speak(self):
        return f"Greetings! I am the Great {self.name}!"
    def cast_spell(self):
        return f"{self.name} casts Fireball!"

c1 = Hero('Elvis')
c2 = Wizard('Merlin')

print(c1.speak())
print(c2.speak())
print(c2.cast_spell())
```
